```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D  
import random
from pandas import read_csv


series=read_csv("reg-inc-data.txt", sep="\t", header=None,parse_dates=True)
series=np.array(series.values)

x=range(0,60,1)





def getXY1(x0,ix):
    return series[x0][ix]






fig, ax = plt.subplots(3, 1,figsize=(15, 15))


ax[0].plot(x, (series[:,3]-series[:,4])/series[:,3], label = "RE generator-doper", linestyle=":")
ax[0].plot(x, (series[:,3]-series[:,5])/series[:,3], label = "RE generator-discriminator", linestyle="-")
ax[0].plot(x, (series[:,4]-series[:,5])/series[:,4], label = "RE doper-discriminator", linestyle="-.")

plt.xlabel('epoch', fontsize=10)
plt.ylabel('relative error', fontsize=10)
plt.title(label="Solar")
ax[0].title.set_text("Solar")
ax[0].legend()
fig.show()




ax[1].plot(x, (series[:,3+5]-series[:,4+5])/series[:,3+5], label = "RE generator-doper", linestyle=":")
ax[1].plot(x, (series[:,3+5]-series[:,5+5])/series[:,3+5], label = "RE generator-discriminator", linestyle="-")
ax[1].plot(x, (series[:,4+5]-series[:,5+5])/series[:,4+5], label = "RE doper-discriminator", linestyle="-.")

plt.xlabel('epoch', fontsize=10)
plt.ylabel('relative error', fontsize=10)
plt.title(label="Univer")
ax[1].title.set_text("Univer")
ax[1].legend()
fig.show()




ax[2].plot(x, (series[:,3+10]-series[:,4+10])/series[:,3+10], label = "RE generator-doper", linestyle=":")
ax[2].plot(x, (series[:,3+10]-series[:,5+10])/series[:,3+10], label = "RE generator-discriminator", linestyle="-")
ax[2].plot(x, (series[:,4+10]-series[:,5+10])/series[:,4+10], label = "RE doper-discriminator", linestyle="-.")

plt.xlabel('epoch', fontsize=10)
plt.ylabel('relative error', fontsize=10)
plt.title(label="India")

ax[2].legend()
fig.show()
```
